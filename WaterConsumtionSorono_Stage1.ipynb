```python
# To  impute the Consumption with the average 
#  value of each Collection depending on the CollectionIndex

counter = 0 
# Imputing each group separately
for lut in df3.LandUseType.unique():
    temp1 = df3.query(f"LandUseType == '{lut}'")
    for ust in temp1.UserType.unique():
        temp2 = temp1.query(f"UserType == '{ust}'")
        for ipd in temp2.InletDiameter.unique():
            temp3 = temp2.query(f"InletDiameter == {ipd}")
            for gvr in temp3.GaugeVendor.unique():              
                temp4 = temp3.query(f"GaugeVendor == '{gvr}'")
                num_nans = (temp4.ImputedReference.isna().sum(),
                            temp4.ImputedConsumption.isna().sum())
                # If the case contains any NaNs
                counter += 1
                if sum(num_nans)>0:

                    temp_frame = temp4#.reset_index().set_index('CurrentDate')
                    # If the Consumption column contains NaNs           
                    if num_nans[1]>0:
                        # take a copy of the column with NaNs
                        temp_col = temp_frame.ImputedConsumption
                        # get the mean
                        temp_mean = temp_col.mean()
                        # getting the slice
                        temp_ref = df3.loc[(df3.LandUseType == lut) & 
                                    (df3.UserType == ust) &
                                    (df3.InletDiameter == ipd) &
                                    (df3.GaugeVendor == gvr) , 
                                    "ImputedConsumption"]
                        # specify the indexes of NaNs
                        idxs=temp_ref.index[np.argwhere(np.isnan(temp_ref.values))].tolist()[0]
                        # Filling with mean
                        for idx in idxs:
                            df3.iloc[idx, list(df3).index('ImputedConsumption')]=temp_mean

                        
                        # Check if it is still has gaps?
                        if temp_ref.isna().sum()>1:
                            print("C", counter, temp_col.isna().sum(), end='; ')
                        else:
                            print("Collection" , counter, " is OK")
```
